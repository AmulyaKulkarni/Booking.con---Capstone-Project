```python
import sqlite3
import pandas as pd
```

```python
db = sqlite3.connect("booking_website_data.db")
cursor = db.cursor()
```

```python
import pandas as pd
file_name = 'final2.csv'
df = pd.read_csv(file_name, encoding='latin1')
```

```python
print(df.columns)
```

```python
table_name = "hotels2"

create_table_query2 = """
CREATE TABLE IF NOT EXISTS new_table (
    id INTEGER PRIMARY KEY,
    hotel_name TEXT,
    freewifi TEXT,
    familyrooms TEXT,
    nonsmokingrooms TEXT,
    restaurant TEXT,
    bar TEXT,
    heating TEXT,
    lift TEXT,
    Cuisine1 TEXT,
    Cuisine2 TEXT,
    Cuisine3 TEXT
)
"""


cursor.execute(create_table_query2)

db.commit()

```

```python
table_name = "new_table"

for index, row in df.iterrows():
    hotel_name = row['Hotel Name']
    freewifi = row['freewifi']
    familyrooms = row['familyrooms']
    nonsmokingrooms = row['nonsmokingrooms']
    restaurant = row['restaurant']
    bar = row['bar']
    heating = row['heating']
    lift = row['lift']
    Cuisine1 = row['Cuisine1']
    Cuisine2 = row['Cuisine2']
    Cuisine3 = row['Cuisine3']

    cursor.execute(f"INSERT INTO {table_name} (hotel_name, freewifi, familyrooms, nonsmokingrooms, restaurant, bar, heating, lift, Cuisine1, Cuisine2, Cuisine3) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)",
                   (hotel_name, freewifi, familyrooms, nonsmokingrooms, restaurant, bar, heating, lift, Cuisine1, Cuisine2, Cuisine3))

db.commit()
```

```python
#1 Retrieve hotels with both a restaurant and a bar
cursor.execute("SELECT * FROM new_table WHERE restaurant = 'Yes' AND bar = 'Yes';")
result1 = cursor.fetchall()

#2 Retrieve hotels with free WiFi and family rooms
cursor.execute("SELECT * FROM new_table WHERE freewifi = 'Yes' AND familyrooms = 'Yes';")
result2 = cursor.fetchall()

#3 Count the occurrences of each breakfast cuisine type
cursor.execute("""
    SELECT Cuisine, COUNT(*) as count
    FROM (
        SELECT Cuisine1 as Cuisine FROM new_table
        UNION ALL
        SELECT Cuisine2 as Cuisine FROM new_table
        UNION ALL
        SELECT Cuisine3 as Cuisine FROM new_table
    )
    GROUP BY Cuisine;
""")
result3 = cursor.fetchall()

#4 Retrieve hotels offering a specific breakfast cuisine (e.g., CONTINENTAL)
cursor.execute("SELECT * FROM new_table WHERE Cuisine1 = 'CONTINENTAL' OR Cuisine2 = 'CONTINENTAL' OR Cuisine3 = 'CONTINENTAL';")
result4 = cursor.fetchall()

#5 Check if a hotel has all amenities (Heating, Lift, and Bar)
cursor.execute("SELECT * FROM new_table WHERE heating = 'Yes' AND lift = 'Yes' AND bar = 'Yes';")
result5 = cursor.fetchall()

print("1 - Hotels with both Restaurant and Bar:")
print(result1)

print("\n2 - Hotels with Free WiFi and Family Rooms:")
print(result2)

print("\n3 - Occurrences of Each Breakfast Cuisine Type:")
print(result3)

print("\n4 - Hotels Offering CONTINENTAL Breakfast:")
print(result4)

print("\n5 - Hotels with All Amenities (Heating, Lift, and Bar):")
print(result5)
```
