```python
import sqlite3
import pandas as pd
```

```python
db = sqlite3.connect("booking_website_data.db")
cursor = db.cursor()
```

```python
import pandas as pd
file_name = 'final1.csv'
df = pd.read_csv(file_name, encoding='latin1')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
table_name = "hotels"

create_table_query = """
CREATE TABLE IF NOT EXISTS hotels (
    id INTEGER PRIMARY KEY,
    hotel_name TEXT,
    city TEXT,
    location TEXT,
    star_rating REAL,
    review_score REAL,
    rating TEXT,
    review_count INTEGER
)
"""


cursor.execute(create_table_query)

db.commit()
```

```python
table_name = "hotels"


for index, row in df.iterrows():
    hotel_name = row['Hotel Name']
    city = row['City']
    location = row['Location']
    star_rating = row['StarRating']
    review_score = row['ReviewScore']
    rating = row['Rating']
    review_count = row['ReviewCount']


    cursor.execute(f"INSERT INTO {table_name} (hotel_name, city, location, star_rating, review_score, rating, review_count) VALUES (?, ?, ?, ?, ?, ?, ?)",
                   (hotel_name, city, location, star_rating, review_score, rating, review_count))

db.commit()
```

```python
# 1
cursor.execute("SELECT * FROM hotels WHERE city = 'Sydney' AND rating > 5;")
result1 = cursor.fetchall()

# 2
cursor.execute("SELECT location, MAX(review_count) as max_reviews FROM hotels;")
result2 = cursor.fetchall()

# 3
cursor.execute("SELECT * FROM hotels ORDER BY rating DESC;")
result3 = cursor.fetchall()

# 4
cursor.execute("SELECT * FROM hotels WHERE star_rating >= 4.0 ORDER BY review_score DESC;")
result4 = cursor.fetchall()

# 5
cursor.execute("SELECT city, COUNT(*) as hotel_count FROM hotels GROUP BY city;")
result5 = cursor.fetchall()

cursor.close()
db.close()


print("1 - Hotels in Sydney with Ratings > 5:")
print(result1)

print("\n2 - Location with the Most Reviews:")
print(result2)

print("\n3 - All Rows Sorted by User Ratings (Descending):")
print(result3)

print("\n4 - Rows with Star Rating >= 4.0 Sorted by Review Score (Descending):")
print(result4)

print("\n5 - Number of Hotels in Each City:")
print(result5)
```
