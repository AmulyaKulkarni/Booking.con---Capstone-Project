```python
db = sqlite3.connect("booking_website_data.db")
cursor = db.cursor()

# 3: Find hotels with user ratings above 5 and that offer free WiFi
query3 = """
SELECT t1."Hotel Name", t1."City", t1."Location", t1."StarRating", t1."ReviewScore", t1."Rating", t1."ReviewCount", t2.freewifi
FROM "table1" AS t1
INNER JOIN "table2" AS t2 ON t1."Hotel Name" = t2."Hotel Name"
WHERE CAST(REPLACE(t1."Rating", 'Very good', '8.5') AS REAL) > 5.0 AND t2.freewifi = 'Yes';

"""
cursor.execute(query3)
result3 = cursor.fetchall()
print(result3)

# 4: Find hotels that offer both family rooms and non-smoking rooms
query4 = """
SELECT t1."Hotel Name"
FROM "table1" AS t1
WHERE t1."Hotel Name" IN (
    SELECT t2."Hotel Name"
    FROM "table2" AS t2
    WHERE t2.familyrooms = 'Yes' AND t2.nonsmokingrooms = 'Yes'
);

"""
cursor.execute(query4)
result4 = cursor.fetchall()
print(result4)

# 5: Calculate the average price for each room type
query5 = """
SELECT t3.RoomType, AVG(t3.Price) AS AveragePrice
FROM "table3" AS t3
GROUP BY t3.RoomType;
"""
cursor.execute(query5)
result5 = cursor.fetchall()
print(result5)

cursor.close()
db.close()
```
