```python
import sqlite3
import pandas as pd

db = sqlite3.connect("booking_website_data.db")
cursor = db.cursor()

file_name = 'final3.csv'
df = pd.read_csv(file_name, encoding='latin1')
```

```python
df.head()
```

```python
table_name = "new_table"

create_table_query = """
CREATE TABLE IF NOT EXISTS new_table (
    id INTEGER PRIMARY KEY,
    HotelName TEXT,
    RoomType TEXT,
    DoubleBed TEXT,
    SingleBed TEXT,
    Price INTEGER
)
"""

cursor.execute(create_table_query)

db.commit()
```

```python
for index, row in df.iterrows():
    data_tuple = tuple(row)
    placeholders = ", ".join("?" * len(row))
    insert_query = f"INSERT INTO {table_name} ({', '.join(row.index)}) VALUES ({placeholders})"
    cursor.execute(insert_query, data_tuple)

db.commit()
```

```python
# To retrieve hotels that offer double bed rooms and sort them by price in descending order

cursor.execute("SELECT * FROM new_table WHERE DoubleBed = 'Yes' ORDER BY Price DESC")
result = cursor.fetchall()

print("Hotels offering Double Bed Rooms sorted by Price (descending order):")
for row in result:
    print(row)
```

```python
# Query to calculate the average price for each room type.

cursor.execute("SELECT RoomType, AVG(Price) AS AveragePrice FROM new_table GROUP BY RoomType")
result = cursor.fetchall()

print("Average Price for Each Room Type:")
for row in result:
    print(row)
```
